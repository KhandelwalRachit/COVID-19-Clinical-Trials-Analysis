# Exploratory Data Analysis of COVID‑19 Clinical TrialsThis notebook performs an extensive Exploratory Data Analysis (EDA) on a dataset ofCOVID‑19 clinical trials collected from ClinicalTrials.gov.Objectives:• Understand distribution of trials by status, phase, and country  • Study demographic participation patterns  • Analyze enrollment statistics  • Investigate time trends in clinical trial launches  • Generate meaningful insights through visualization

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## 2. Load Dataset

```python
df = pd.read_csv("C:\\Users\\Divyanshu\\OneDrive\\Desktop\\COVID Project\\COVID clinical trials.csv")

df.head()
```

## 5. Missing Values Analysis

```python
missing_data = df.isnull().mean()*100
missing_data.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
missing_data.sort_values(ascending=False).plot(kind='bar')

plt.title('Percentage of Missing Values in Dataset')
plt.ylabel('Percentage')
plt.show()
```

## 6. Drop Columns with Extremely High Missing Values

```python
df.drop(['Results First Posted','Study Documents'], axis=1, inplace=True)

df.shape
```

## 7. Duplicate Rows Check

```python
df.duplicated().sum()
```

## 8. Feature Engineering – Extract Country

```python
countries = [str(df.Locations.iloc[i]).split(',')[-1] for i in range(df.shape[0])]

df['Country'] = countries

df['Country'].value_counts().head()
```

## 9. Top Countries Conducting Clinical Trials

```python
top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')

plt.title('Top Countries Conducting COVID Clinical Trials')
plt.xlabel('Country')
plt.ylabel('Number of Trials')
plt.show()
```

## 10. Clinical Trial Status Distribution

```python
status = df['Status'].value_counts()

plt.figure(figsize=(10,6))
status.plot(kind='bar')

plt.title('Status of Clinical Trials')
plt.xlabel('Status')
plt.ylabel('Number of Trials')
plt.show()
```

## 11. Gender Distribution

```python
gender = df['Gender'].value_counts()

plt.figure(figsize=(8,5))
gender.plot(kind='bar')

plt.title('Gender Eligibility Distribution')
plt.show()
```

## 12. Study Type Distribution

```python
study_type = df['Study Type'].value_counts()

plt.figure(figsize=(8,5))
study_type.plot(kind='bar')

plt.title('Study Type Distribution')
plt.show()
```

## 13. Phase Distribution

```python
phase = df['Phases'].value_counts()

plt.figure(figsize=(8,5))
phase.plot(kind='bar')

plt.title('Clinical Trial Phase Distribution')
plt.show()
```

## 14. Enrollment Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Enrollment'], bins=50)

plt.title('Enrollment Distribution')
plt.show()
```

## 15. Handling Missing Data

```python
categorical_features = df.select_dtypes(include='object').columns

for feature in categorical_features:
    df[feature] = df[feature].fillna(f'Missing {feature}')
```

```python
median_enrollment = df['Enrollment'].median()

df['Enrollment'] = df['Enrollment'].fillna(median_enrollment)
```

## 16. Time Series Analysis – Trials Started Over Time

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

df['Start Month'] = df['Start Date'].dt.to_period('M')

monthly_trials = df['Start Month'].value_counts().sort_index()

plt.figure(figsize=(12,6))
monthly_trials.plot(kind='line')

plt.title('Clinical Trials Started Over Time')
plt.ylabel('Number of Trials')
plt.show()
```

## 17. Phase vs Status Relationship

```python
phase_status = pd.crosstab(df['Status'], df['Phases'])

phase_status.plot(kind='bar', stacked=True, figsize=(12,6))

plt.title('Status vs Phase of Clinical Trials')
plt.show()
```
